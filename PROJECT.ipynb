# VISUALIZATION 


# 1.Calculating MEAN of release_year


```python
release_year_counts = df['release_year'].value_counts().sort_index()
mean_count = release_year_counts.mean()
plt.figure(figsize=(10, 8))
plt.barh(release_year_counts.index, release_year_counts.values, color='Red')
plt.axvline(x=mean_count, color='black', linestyle='--', label=f'Mean Count: {mean_count:.2f}')
plt.title('Number of Releases per Year on Netflix')
plt.xlabel('Number_of_Releases')
plt.ylabel('Release_Year')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
type_counts = df['type'].value_counts()
plt.figure(figsize=(5, 4.5))  # Set figure size
plt.bar(type_counts.index, type_counts.values, color=['Red', 'Black'])
plt.title('Netflix Shows Distribution')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
labels=["Movie","TV Show"]
size=df["type"].value_counts()
colors = ['#E50914', '#000000']
explode=[0,0.1]
plt.figure(figsize=(4, 4))
plt.rcParams["figure.figsize"]=(9,9)
plt.pie(size, labels=labels, colors=colors, autopct='%1.1f%%', explode=explode, shadow=True,textprops=dict(color="white", weight="bold", fontsize=11), startangle=90)
plt.title("Movies vs. TV Shows on Netflix")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='rating', order=df['rating'].value_counts().index, palette='viridis')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.title('Distribution of Netflix Ratings')
plt.xticks(rotation=45)  
plt.show()
```

```python
#6.Show the Top 5 Netflix country distribution separated by type of release (i.e Movies, TV shows), visualizing via Bar Graph.#
# Count the number of movies and TV shows per country
country_type_counts = df.groupby(['country', 'type']).size().reset_index(name='count')

# Get the top 5 countries with the highest number of shows/movies
top_countries = country_type_counts.groupby('country')['count'].sum().nlargest(5).index
filtered_data = country_type_counts[country_type_counts['country'].isin(top_countries)]

# Create a bar plot
plt.figure(figsize=(12, 8))
sns.barplot(data=filtered_data, x='country', y='count', hue='type', palette='viridis')
plt.xlabel('Country')
plt.ylabel('Count')
plt.title('Top 5 Countries by Type of Release on Netflix')
plt.xticks(rotation=45)  # Rotate x-axis labels if needed
plt.legend(title='Type')
plt.show()
```

```python
#7.Show the top 10 genres of movies and also tv shows.
df['genre'] = df['genre'].str.split(',')
df = df.explode('genre')

# Filter data for movies and TV shows
movies_df = df[df['type'] == 'Movie']
tv_shows_df = df[df['type'] == 'TV Show']

# Get top 10 genres for movies
top_10_movie_genres = movies_df['genre'].value_counts().head(10)

# Get top 10 genres for TV shows
top_10_tv_genres = tv_shows_df['genre'].value_counts().head(10)

# Plotting
fig, axes = plt.subplots(2, 1, figsize=(12, 12))

# Top 10 movie genres
top_10_movie_genres.plot(kind='bar', ax=axes[0], color='Red')
axes[0].set_title('Top 10 Movie Genres on Netflix')
axes[0].set_xlabel('Genre')
axes[0].set_ylabel('Count')
axes[0].set_xticklabels(top_10_movie_genres.index, rotation=45)

# Top 10 TV show genres
top_10_tv_genres.plot(kind='bar', ax=axes[1], color='Black')
axes[1].set_title('Top 10 TV Show Genres on Netflix')
axes[1].set_xlabel('Genre')
axes[1].set_ylabel('Count')
axes[1].set_xticklabels(top_10_tv_genres.index, rotation=45)

plt.tight_layout()
plt.show()
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['month'] = df['date_added'].dt.month
df['year'] = df['date_added'].dt.year
heatmap_data = df.pivot_table(index='year', columns='month', aggfunc='size', fill_value=0)
plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data, cmap='Blues', annot=True, fmt='d')
plt.title('Best Time to Release a Movie/TV Show (Heatmap)')
plt.show()
```
